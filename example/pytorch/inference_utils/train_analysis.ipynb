# Analysis of real-time inference impact on training time & accuracy

```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
```

```python
pickle_file = 'trani-time-24-no-inf.pkl'
with open(pickle_file, 'r') as time_and_accu_file:
    time_and_accu = pickle.load(time_and_accu_file)
train_time = time_and_accu[0]
train_accu = [i.item() for i in time_and_accu[1]]

chart1, ax1 = plt.subplots()
ax1.plot(train_time)
plt.title('training time')

chart2, ax2 = plt.subplots()
ax2.plot(train_accu)
plt.title('training accu')
```

```python
pickle_file = 'trani-time-24.pkl'
with open(pickle_file, 'r') as time_and_accu_file:
    time_and_accu = pickle.load(time_and_accu_file)
train_time = time_and_accu[0]
train_accu = [i.item() for i in time_and_accu[1]]

chart1, ax1 = plt.subplots()
ax1.plot(train_time)
plt.title('training time')
# plt.savefig('train_time_24_eval_only.png')

chart2, ax2 = plt.subplots()
ax2.plot(train_accu)
plt.title('training accu')
# plt.savefig('train_accu_24_eval_only.png')
```

```python
def all_stats(a):
    return ['{:.3f}'.format(max(a)), '{:.3f}'.format(statistics.mean(a)), '{:.3f}'.format(statistics.median(a)),
            '{:.3f}'.format(statistics.stdev(a)), '{:.3f}'.format(min(a))]
print('label ', 'max', 'mean', 'median', 'std', 'min')
print 'train_accu', all_stats(train_accu)
print 'train_time', all_stats(train_time)
```
